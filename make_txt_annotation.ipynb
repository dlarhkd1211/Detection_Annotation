```python
import xml.etree.ElementTree as ET
from xml.etree.ElementTree import Element, dump, ElementTree
from glob import glob
import os
import re
from PIL import Image
```

```python
path = glob('C:\\pet\\exam\\oxford_pet\\annotations\\xmls\\*')
```

```python
file_name = []
for i in path:
    class_name = re.sub('_\d+', '', i.split('\\')[-1].replace('.xml',''))
#     print(class_name)
    if class_name[0].islower():
        continue
    else:
        file_name.append(i)
    
```

```python
file_name
```

```python
o_path = 'C:\\Users\\localley\\Desktop\\images'
for i in file_name:
    doc = ET.parse(i)
    root = doc.getroot()
    f = open(o_path + '\\' + i.split('\\')[-1].replace('xml','txt'), 'w')
    for object in root.iter("size"):
        width = int(object.findtext("width"))
        height = int(object.findtext("height"))

    for object in root.iter("object"):
        xmin = int(object.find("bndbox").findtext("xmin"))
        xmax = int(object.find("bndbox").findtext("xmax"))
        ymin = int(object.find("bndbox").findtext("ymin"))
        ymax = int(object.find("bndbox").findtext("ymax"))

        
        classes = '2'
        f.write(str(classes))
        f.write(' ')
        f.write(str(((xmin + xmax)//2)/width))
        f.write(' ')
        f.write(str(((ymin + ymax)//2)/height))
        f.write(' ')
        f.write(str((xmax - xmin)/width))
        f.write(' ')
        f.write(str((ymax - ymin)/height))
        f.write('\n')
```
